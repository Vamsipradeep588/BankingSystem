```python
"""
PlatinumRx Assignment | Phase 3 – Python
Script 1: Time Converter
========================
Given a number of minutes, convert it into a human-readable string.

Examples:
    130  →  "2 hrs 10 minutes"
    110  →  "1 hr 50 minutes"
    60   →  "1 hr 0 minutes"
    45   →  "0 hrs 45 minutes"
"""


def convert_minutes(total_minutes: int) -> str:
    """
    Convert an integer number of minutes into a human-readable string.

    Parameters
    ----------
    total_minutes : int
        Total duration in minutes (must be >= 0).

    Returns
    -------
    str
        Formatted string like "2 hrs 10 minutes".
    """
    if total_minutes < 0:
        return "Invalid input: minutes cannot be negative."

    hours           = total_minutes // 60       # integer division → whole hours
    remaining_mins  = total_minutes  % 60       # modulo → leftover minutes

    # Pluralisation for "hr" / "hrs"
    hr_label  = "hr"  if hours == 1 else "hrs"
    min_label = "minute" if remaining_mins == 1 else "minutes"

    return f"{hours} {hr_label} {remaining_mins} {min_label}"


# ── Main: test with sample inputs ───────────────────────────────────────────

if __name__ == "__main__":
    test_cases = [130, 110, 60, 45, 0, 1, 61, 999]

    print("=" * 40)
    print("  Time Converter – Results")
    print("=" * 40)
    for minutes in test_cases:
        result = convert_minutes(minutes)
        print(f"  {minutes:>5} minutes  →  {result}")
    print("=" * 40)

    # ── Interactive mode ─────────────────────────────────────
    print("\nEnter a number of minutes to convert (or 'q' to quit):")
    while True:
        user_input = input("  Minutes: ").strip()
        if user_input.lower() == 'q':
            print("Goodbye!")
            break
        try:
            mins = int(user_input)
            print(f"  Result : {convert_minutes(mins)}\n")
        except ValueError:
            print("  Please enter a valid integer.\n")
```

```python
"""
PlatinumRx Assignment | Phase 3 – Python
Script 2: Remove Duplicate Characters
======================================
Given a string, remove all duplicate characters using a loop
and print the resulting unique string (preserving first-occurrence order).

Examples:
    "programming"  →  "progamin"
    "hello world"  →  "helo wrd"
    "aabbcc"       →  "abc"
"""


def remove_duplicates(input_string: str) -> str:
    """
    Remove duplicate characters from a string using a loop.
    The first occurrence of every character is kept;
    subsequent duplicates are discarded.

    Parameters
    ----------
    input_string : str
        The raw string that may contain duplicate characters.

    Returns
    -------
    str
        A new string with each character appearing exactly once,
        in the order they first appeared.
    """
    result = ""                         # start with an empty string

    for char in input_string:          # loop through every character
        if char not in result:          # only add if not already seen
            result += char              # append first-occurrence characters

    return result


# ── Main: test with sample inputs ───────────────────────────────────────────

if __name__ == "__main__":
    test_cases = [
        "programming",
        "hello world",
        "aabbcc",
        "PlatinumRx",
        "data analyst",
        "mississippi",
        "",                             # edge case: empty string
        "a",                            # edge case: single character
        "aaaa",                         # edge case: all same characters
    ]

    print("=" * 50)
    print("  Remove Duplicates – Results")
    print("=" * 50)
    for s in test_cases:
        unique = remove_duplicates(s)
        print(f"  Input  : {repr(s)}")
        print(f"  Output : {repr(unique)}")
        print("-" * 50)

    # ── Interactive mode ─────────────────────────────────────
    print("\nEnter a string to remove duplicates (or 'q' to quit):")
    while True:
        user_input = input("  String : ")
        if user_input.lower() == 'q':
            print("Goodbye!")
            break
        print(f"  Result : {remove_duplicates(user_input)}\n")
```
